```python
import pandas as pd
import numpy as np
import re
```

```python
url = 'https://raw.githubusercontent.com/Theodore-Ptr/Hackathon/main/full_data.csv'
df = pd.read_csv(url, sep='|',index_col = 0)
df = df.drop_duplicates()
```

```python
for idx, item in enumerate(df.injection_params.unique()):
    print(str(idx) + '  ' + item)
```

```python
regex = re.compile(r'(^|(, )|(; )|(\. )|(  ))(\d+)( |)(cvd)*(корунд синт.)*(куб.)*(((?!\d|(Круг)|(бесцв\.)|(Кр )|(лондон)|(багет)|(Кр-)|(квадрат)|(кр)|(овал)|(культ\.)|(круг)|(недраг)|\;|(иск.)|(sw)|(swarovski)|(лск)|(сваровски)|(Swarovski)|(б\\ц)|(б\/ц)|(бесцветный)|(белый)|(черный)|\/|(нат\. ов)|(ов-)).)*)')

count_gr = 5
name_gr = 10

def format_names(name):
    name = name.strip().lower()
    name = re.sub(r'(бриллиант)|(брилл\.)|(бр)', 'бриллиант', name)
    name = re.sub(r'фианит сапфир', 'фианит', name)
    name = re.sub(r'бп-', 'бп', name)
    return name


def parse_name(string):
    string = string.lower()
    stones = []
    for p in re.findall(regex, string):
        if not p[name_gr] or p[name_gr] == ',':
            continue
        stones += [{'name':format_names(p[name_gr]), 'count':p[count_gr]}]
    return stones
    

data = []
line = []
ring = 0

for item in df.injection_params.unique():
    for s in parse_name(item):
        line = [ring, s['name'], s['count']]
    
        data += [line]
    ring += 1
new_df = pd.DataFrame(data, columns = ['ring id', 'Name', 'Count'])

#with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
#    print(new_df)

new_df.Name.drop_duplicates().tolist()

print('ok')
```

```python
add_col = new_df.Name.drop_duplicates().tolist()

cols = {}

for i in add_col:
    cols[i] = []

c = 0
for idx, row in df.iterrows():
    for i in add_col:
        cols[i] += [0]
    
    for stone in parse_name(row['injection_params']):
        cols[stone['name']][-1] += int(stone['count'])
    
    c += 1

for name, l in cols.items():
    df[name] = l
```
